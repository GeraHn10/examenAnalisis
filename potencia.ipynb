```
import matplotlib.pyplot as plt
import numpy as np
import time as t

def power(x,n):
   if n==0:
     return 1

   if (n&1) !=0:
      return x * power(x*x, n>>1)
   else:
      return power(x*x, n>>1)


times = {}


def test(x,p):
  start_time = t.time()
  print(power(x,p))
  n=p
  times[n]= t.time() - start_time
  print("--- %s seconds ---" % (times[n]))



def plot(times):
  x = sorted(times.items())
  plt.plot(x)
  plt.show()
  print(times)

base=3
potencia=100
for i in range(100):
  test(base,potencia)
  potencia=potencia-1

plot(times)    
```

#Analisis de complejidad
El algoritmo da como resultante un complejidad de O(LOG N)
Siendo altamente efectivo por su trabajo a nivel de bits y recursividad
